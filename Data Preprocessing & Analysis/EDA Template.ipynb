# Data Preparation (EDA)

## Import the necessary packages

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tkinter import Tk
from tkinter.filedialog import askopenfilename
```

## Load the dataset

```python
# Prompt to choose a file
Tk().withdraw()  # We don't want a full GUI, so keep the root window from appearing
file_path = askopenfilename()  # Show an "Open" dialog box and return the path to the selected file

# Load the dataset
df = pd.read_csv(file_path)
```

## Understand the data, get a snapshot and see the data types

```python
# Inspect the structure of the dataset
print(df.head())
print(df.info())
print(df.describe())

# Identify missing values
missing_values = df.isnull().sum()
print(missing_values)

# Display rows with missing values
rows_with_missing_values = df[df.isnull().any(axis=1)]
print(rows_with_missing_values)
```

## Data Cleaning

```python
# Data Cleaning
# Remove anomalous values
df = df[df['max_heart_rate'] >= 0]
df = df[df['age'] >= 0]
df = df[df['non_anginal_pain'] >= 0]

# Change datatype of age from float to integer
df['age'] = df['age'].astype(int)

# Fill missing values in angina_level with 'n/a'
df['angina_level'] = df['angina_level'].fillna('n/a')

# One-hot encoding for angina_level
df = pd.get_dummies(df, columns=['angina_level'], drop_first=False)
```

# Handling Missing Data

```python
# Identify missing values
missing_values = df.isnull().sum()
print(missing_values)

print(df.head())
# If any is missing, drop the row
# df = df.dropna()
```

# Transforming

```python
# Convert categorical variables to numerical values
df = pd.get_dummies(df, drop_first=True)
```

# Summarizing and Visualization

```python
# Visualize distributions
df.hist(bins=30, figsize=(20, 15))
plt.show()

# Visualize relationships
sns.pairplot(df)
plt.show()
```

## Model Evaluation

<summary>Describe possible ways to evaluate model performance.</summary>

<div class="alert alert-block alert-info">

### Evaluation Metrics
- **Accuracy**: The ratio of correctly predicted instances to the total instances. It is a good measure when the classes are well balanced.
- **Precision**: The ratio of correctly predicted positive observations to the total predicted positives. It is useful when the cost of false positives is high.
- **Recall (Sensitivity)**: The ratio of correctly predicted positive observations to all the observations in the actual class. It is useful when the cost of false negatives is high.
- **F1-Score**: The weighted average of Precision and Recall. It is useful when you need a balance between Precision and Recall.
- **ROC-AUC Score**: The area under the Receiver Operating Characteristic curve. It provides an aggregate measure of performance across all classification thresholds.

### Importance
- **Accuracy**: Important for getting a general sense of model performance but can be misleading with imbalanced datasets.
- **Precision**: Important in scenarios where false positives are costly, such as spam detection.
- **Recall**: Important in scenarios where false negatives are costly, such as medical diagnosis.
- **F1-Score**: Provides a balance between Precision and Recall, useful when you need to consider both metrics.
- **ROC-AUC Score**: Useful for comparing models and understanding the trade-off between true positive rate and false positive rate.

<div class="alert alert-block alert-warning">
<b>Example:</b> For a medical diagnosis problem, precision and recall might be more important than accuracy because false positives and false negatives have different implications.
</div>
